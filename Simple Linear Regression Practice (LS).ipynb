This notebook deal with using the least square method of simple linear regression  for analysis and prediction in python.s

1. The first step involves importing all the neccesary libraries for computation.

```python
import pandas as pd #This would help us analyse better since we need our dataset in a dataframe to do computations
import numpy as np # Numpy helps us comppute individual arrays faster and efficiently
import matplotlib.pyplot as plt # This library helps to plot all the chart that is required of us
from matplotlib import rc
from sklearn.linear_model import LinearRegression
```

2. The next step is importing our dataset and storing it in a dataframe.

```python
df = pd.read_csv("C:/Users/Christopher.Ezereony/Downloads/archive/test.csv")
```

3. Plot the scatter plot and line of best fit

```python
#rename the linear regression model making it easy to make use of when you need to.
model = LinearRegression()

# create variabele to help shape your dataset
X = df[['x']]
y = df['y']

#fit your model
model.fit(X,y)

#calculate the slope and intercept
m = model.coef_[0]
c = model.intercept_

# calculate predicted values
prediction = model.predict(X)

#
# plot your scatter diagram
plt.scatter(df.x,df.y, color='green')
plt.title("Plot of X against Y")
plt.ylabel("Y dataset")
plt.xlabel("X dataset")
plt.plot(X,prediction, color='red')

print("Slope:\t\t", m)
print("Intercept:\t", c)
```

4. Calculate the correlation coefficient and the mean squared error

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

mse = mean_squared_error(y, prediction)
r_squared = r2_score(y, prediction)
mae = mean_absolute_error(y, prediction)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r_squared:.2f}')
print(f'Mean Absolute Error: {mae:.2f}')
```

Interpretation: With a high correlation coeficient we can say that we observe a strong positive relationship between the predicted values and the actual values. This model is expected to predict optimally under normal circumstance and with exemption of unforseen future interference.
